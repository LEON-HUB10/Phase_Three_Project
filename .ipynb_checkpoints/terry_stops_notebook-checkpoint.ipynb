# Terry Traffic Stops Classification Project

## Business Understanding
The Terry Stops allows law enforcement officers to briefly detain individuals based on reasonable suspicion, without necessarily having evidence to justify a full arrest. While these stops are legally sanctioned, they raise questions about how arrests are decided after such stops and whether factors like weapons presence, time of the day, call type or race and gender influence the outcome

## Problem Statement
The aim is to predict whether a Terry Stop results in an arrest based on situational factors like time, type of call, presence of weapons or officer attributes are subject to demographics.

Understanding these patterns can:
* Help improve officer decision-making transparency.
* Identify potential biases in law enforcement practices.
* Support data-driven policy recommendations for fair and equitable policing.

## Objectives
* To build a predictive model that classifies whether a Terry Stop results in an arrest.
* To identify key drivers of arrest during Terry Stops.
* Explore Whether subject race, officer race or gender significantly affect arrest likelihood.
* Highlight potential patterns of bias or disparities for further policy and ethics discussions.

## Metrics of Success
* To achieve an accuracy score of 75% and above.
* To achieve an f1-score of 75% and above.

## Data Understanding
* The dataset has 24 columns of which 4 are numerical and 20 are categorical.
* The dataset has identifiers such as _id, terry_stop_id, subject_id, go_sc_num and officer_id.
* It has event information such as initial_call_type, final_call_type, call_type, frisk_flag, stop_resolution and weapon_type.
* It has Time and location attributes such as reported_time, reported_date, precinct, beat and  sector.
* It has officer attributes such as officer_gender, officer_race, officer_squad and officer_yob.
* It has subject attributes such as subject_percieved_race, subject_perceived_gender and subject_age_group.
* The target variable is arrest_flag which is either Yes or No (Y/N).

## Hypothesis Testing

#### Hypothesis Test1: Subject perceived race against Arrest flag
Null Hypothesis: There is no significant relationship between subject_perceived_race and arrest_flag.

Alternative Hypothesis: There is a significant relationship between subject_perceived_race and arrest_flag.

```python
# Creating contingency table
race_arrest_ct = pd.crosstab(data1['subject_perceived_race'], data1['arrest_flag'])

# Running the chi-square test
chi2, p, dof, expected = chi2_contingency(race_arrest_ct)

print('Chi-square Statistic:', chi2)
print('Degrees of Freedom:', dof)
print('P-value:', p)
```

Observations: The p-value is less than 0.05 level of significance therefore we reject the null hypothesis and conclude that there is a significant relationship between subject_perceived_race and arrest_flag.

#### Hypothesis Test2: Subject perceived gender against Arrest flag
Null Hypothesis: There is no significant relationship between subject_perceived_gender and arrest_flag.

Alternative Hypothesis: There is a significant relationship between subject_perceived_gender and arrest_flag.

```python
# Creating contingency table
gender_arrest_ct = pd.crosstab(data1['subject_perceived_gender'], data1['arrest_flag'])

# Running the chi-square test
chi2, p, dof, expected = chi2_contingency(gender_arrest_ct)

print('Chi-square Statistic:', chi2)
print('Degrees of Freedom:', dof)
print('P-value:', p)
```

Observations: The p-value is less than 0.05 therefore we reject the null hypothesis and conclude that there is a significant relationship between subject_perceived_gender and arrest flag.

Arrest rates are not evenly distributed across gender categories suggesting the subject gender plays a role in determining arrest outcomes.

#### Hypothesis Test3: Officer race against Arrest flag
Null Hypothesis: There is no significant relationship between officer_race and arrest_flag.

Alternative Hypothesis: There is a significant relationship between officer_race and arrest_flag.

```python
# Creating contingency table
officer_race_arrest_ct = pd.crosstab(data1['officer_race'], data1['arrest_flag'])

# Running the chi-square test
chi2, p, dof, expected = chi2_contingency(officer_race_arrest_ct)

print('Chi-square Statistic:', chi2)
print('Degrees of Freedom:', dof)
print('P-value:', p)
```

Observations: The p-value is less than 0.05 therefore we reject the null hypothesis and conclude that there is a significant relationship between officer_race and arrest_flag.

The distribution of arrests during terry stops varies significantly across officer race categories. It might suggest patterns that may reflect systematic differences in officer decision making.

#### Hypothesis Test4: Age group against Arrest flag
Null Hypothesis: There is no significant relationship between subject_age_group and arrest_flag.

Alternative Hypothesis: There is a significant relationship between subject_age_group and arrest_flag.

```python
# Creating contingency table
age_group_arrest_ct = pd.crosstab(data1['subject_age_group'], data1['arrest_flag'])

# Running the chi-square test
chi2, p, dof, expected = chi2_contingency(age_group_arrest_ct)

print('Chi-square Statistic:', chi2)
print('Degrees of Freedom:', dof)
print('P-value:', p)
```

Observations: the p-value is less than 0.05 therefore we reject the null hypothesis and conclude that there is a significant relationship between subject_age_group and arrest_flag.

This suggests that certain age groups have higher or less arrest than other age groups.

#### Hypothesis Test5: Frisk Flag against Arrest flag
Null Hypothesis: There is no significant relationship between frisk_flag and arrest_flag.

Alternative Hypothesis: There is a significant relationship between frisk_flag and arrest_flag.

```python
# Creating contingency table
frisk_arrest_ct = pd.crosstab(data1['frisk_flag'], data1['arrest_flag'])

# Running the chi-square test
chi2, p, dof, expected = chi2_contingency(frisk_arrest_ct)

print('Chi-square Statistic:', chi2)
print('Degrees of Freedom:', dof)
print('P-value:', p)
```

Observations: The p-value is less than 0.05 therefore we reject the null hypothesis and conclude that there is a significant relationship between frisk_flag and arrest_flag.

This suggests that if a frisk is conducted, the probability of an arrest is higher compared to when there is no frisking.
